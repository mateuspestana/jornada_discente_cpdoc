## Definição e história

Python é uma linguagem orientada a objetos, de alto nível, que assim como o R, que também é muito utilizado no meio acadêmico, nos permite desenvolver análises, relatórios, modelos estatísticos, sites, dentre outras coisas. 

Python possui mais bibliotecas e é uma linguagem um pouco mais versátil, não tendo um foco mais específico. O R, apesar de ter um foco específico em estatística, nos permite fazer outras coisas. O Python nunca teve um foco específico, o que garantiu o desenvolvimento da linguagem em diversas áreas: criação de jogos, softwares, estatística, redes neurais, visualizações, e até mesmo sistemas. 

O Python foi criado por Guido von Rossum, em 1991, e vem se aprimorando a cada dia, estando na versão 3.10.

Python é considerada uma linguagem fácil, mesmo para quem não tem background de programação, mas não por isso deve ser entendida como uma linguagem superior ou melhor que outras. Um bom pesquisador resolve problemas, independente da lingugem que usa. O ideal é saber ao máximo, dado que alguns problemas podem ser mais fáceis (ou mais diretos) no R, e outros no Python, por exemplo, ou mesmo na linguagem Julia.

```python
class Calcula:
  def soma(a, b):
    return a + b
  
  def subtracao(a, b):
    return a - b

  def multiplicacao(a, b):
    return a * b

  def divisao(a, b):
    return a/b

print(Calcula.soma(4, 2))
print(Calcula.subtracao(10, 5))
print(Calcula.multiplicacao(2, 5))
print(Calcula.divisao(10, 2))
```

```python
class Estudante(object):
    def __init__(self, nome, idade, genero, nivel, programa, notas=None):
        self.nome = nome
        self.idade = idade
        self.genero = genero
        self.nivel = nivel
        self.programa = programa
        self.notas = notas or {}

    def setNota(self, curso, nota):
        self.notas[curso] = nota

    def getNota(self, curso):
        return self.notas[curso]

    def getCR(self):
        return sum(self.notas.values())/len(self.notas)

matheus = Estudante("Matheus", 29, "M", 'Doutorado',  'Ciência Política', {"TP 1":9.5, "TP 2":9.0, "TP 3":7.0})
julia = Estudante("Julia", 26, "F", 'Doutorado', 'Ciência Política', {"TP 3":10.0, 'Lego 1':8.0})

# Now we can get to the grades easily
print(matheus.getCR())
print(julia.getCR())
```

Podemos gerar uma função que baixa todos os IDs, e depois as informações completas de cada deputado:

```python
def baixaListaDeputados():
  url = 'https://dadosabertos.camara.leg.br/api/v2/deputados?ordem=ASC&ordenarPor=nome'
  r = requests.get(url)
  r.encoding='utf-8'
  j = r.json()
  return pd.DataFrame.from_dict(j['dados'])

def baixaInfoDeputado():
  url_deputado = 'https://dadosabertos.camara.leg.br/api/v2/deputados/'
  lista_deputados = baixaListaDeputados()['id']
  df = pd.DataFrame()
  for id_deputado in lista_deputados[0:20]:
    print(f'Baixando deputado {id_deputado}')
    r = requests.get(url_deputado + str(id_deputado))
    r.encoding='utf-8'
    j = r.json()
    df = pd.concat([df, pd.DataFrame(pd.json_normalize(j['dados']))])
  return df

baixaInfoDeputado()
  
```

### Análise de sentimento

```python
from transformers import pipeline
import pandas as pd
from tqdm.auto import tqdm
tqdm.pandas()


model_path = "cardiffnlp/twitter-xlm-roberta-base-sentiment"
sentiment_task = pipeline("sentiment-analysis", model=model_path, tokenizer=model_path)
```

```python
valencia = sentiment_task('Coloque aqui uma frase qualquer')
print(valencia)
```

```python
sentiment_task(["Foi uma tragédia que afetou a todos", 'Todos ficaram contentes em saber do resultado', 'Os alunos acharam a aula curiosa'])
```

```python
def classificador(texto):
  sentimento = sentiment_task(texto)[0]['label']
  return sentimento
```

```python
classificador('Saio da vida para entrar na História')
```

### Classificação em temas

```python
biden_tweets = pd.read_csv('https://raw.githubusercontent.com/aratrikchandra/US_Presedential_Election2020_Twitter_Sentiment_Analysis/main/Extraction/Biden_No_Retweet_Full_Text.csv')
biden_tweets.tail(100)
```

```python
classificador_tema = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
temas = ['corrupção', 'economia', 'política', 'pandemia', 'imigrantes', 'clima', 'fraude', 'eleição']
```

```python
print(classificador_tema('The latino people deserve attention', temas))
print(classificador_tema('O povo latino merece atenção', temas))
```

```python
def classifica_tema(texto, temas):
  tema = classificador_tema(texto, temas)
  return tema['labels'][0]

classifica_tema('O problema do país está na falta de emprego para o povo', temas)
```

```python
df_biden = biden_tweets[['User', 'Tweet']].sample(15)

df_biden['Tema'] = df_biden['Tweet'].progress_apply(lambda texto: classifica_tema(texto, temas))
```
